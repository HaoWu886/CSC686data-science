```r
# create the dataframe and read the file
library(readr)
companyDF <- read_csv('https://raw.githubusercontent.com/bforoura/AML/master/Company_data.csv')

companyDF
```

```r
# print out the top 10 row
head(companyDF,10)
```

```r
# describe the data set
summary(companyDF)
```

```r
# trying to compute the standard correlation coefficient between attributes
# to determine which parameter correlation with Sales




cor.test(companyDF$TV, companyDF$Sales, method=c("pearson"))
cor.test(companyDF$Radio, companyDF$Sales, method=c("pearson"))
cor.test(companyDF$Newspaper, companyDF$Sales, method=c("pearson"))

# therefore, Tv has the highest correlation with sales
```

```r
# use boxplot to draw a picture
boxplot(companyDF)
```

```r
# plotting histogram and  kernel density estimates 

hist(companyDF$Sales, col = "purple" , prob = TRUE)
lines(density(companyDF$Sales), col = "blue")

hist(companyDF$TV, col = "red" , prob = TRUE)
lines(density(companyDF$TV), col = "blue")

hist(companyDF$Radio, col = "orange" , prob = TRUE)
lines(density(companyDF$Radio), col = "blue")

hist(companyDF$Newspaper, col = "green" , prob = TRUE)
lines(density(companyDF$Newspaper), col = "blue")
```

```r
# check whether it contains null or not
sum(is.na(companyDF))
```

```r
# create a scatter plot to compare TV Radio, newspaper with sales

plot( x = companyDF$TV, y = companyDF$Sales, col = "blue")
plot( x = companyDF$Radio, y = companyDF$Sales, col = "red")
plot( x = companyDF$Newspaper, y = companyDF$Sales, col = "green")
```

```r
# use the data set for training

# create a linear model and use "~" to make connection with TV and sales

model1 <- lm (companyDF$Sales ~ companyDF$TV)


#  print out the intercept and slope values for model not trained
model1
```

```r
# make predictions using the entire data set

# calculate the MSE 
mse <- mean(model1$residuals^2) 

print (mse)
# calculate the R2

summary(model1)$r.squared
```

```r
# split the data and 70% in train, 30% in test
# set seed as 200

library("caret")
set.seed(1)
indexes <- createDataPartition(companyDF$Sales, p = 0.70, list=FALSE, times=1 )
train <- companyDF[indexes,]
test <- companyDF[-indexes,]
```

```r
# 70 percent of 200 for training 
dim(train)
```

```r
# 30 percent of 200 for testing
dim(test)
```

```r
# create a new linear model and use "~" to make connection with TV and sales

model2 <- lm(Sales ~ TV, data = train)


#  print out the coefficients
cat("intercept", "slope", model2$coefficients)

# so we can see the intercept is 7.28, and the slope is 0.053
```

```r
# make prediction 
p <- predict(model2, test)

summary(p)

mean(model2$residuals^2)

summary(model2)$r.squared
```

```r
# plot the best fit linear models

coefs <- coef(model2)
coefs2 <- coef(model1)
plot(Sales ~ TV, train, pch = 15, col = "blue")
abline(coefs, col = "red", lwd=4,)
abline(coefs2, col = "green", lwd=4,)
legend(2,4,legend=c("this trained on 70% and tested on 30% (r2 = 0.79329)"," model trained MSE = 5.62270"))
```

```r
# create a linear model and with Radio and sales

model3 <- lm (companyDF$Sales ~ companyDF$Radio)


#  print out the intercept and slope values for model not trained
model3
```

```r
# make predictions using the entire data set

# calculate the MSE 
mse <- mean(model3$residuals^2) 

print (mse)
# calculate the R2

summary(model3)$r.squared
```

```r
# plot the best fit linear models

coefs <- coef(model3)

plot(Sales ~ Radio, train, pch = 15, col = "blue")
abline(coefs, col = "red", lwd=4,)

legend(2,4,legend=c("this (r2 = 0.1222)"," model trained MSE = 24.38405"))
```

```r
# create a linear model and with Radio and sales

model4 <- lm (companyDF$Sales ~ companyDF$Newspaper)


#  print out the intercept and slope values for model not trained
model4
```

```r
# make predictions using the entire data set

# calculate the MSE 
mse <- mean(model4$residuals^2) 

print (mse)
# calculate the R2

summary(model4)$r.squared
```

```r
# plot the best fit linear models

coefs <- coef(model4)

plot(Sales ~ Newspaper, train, pch = 15, col = "blue")
abline(coefs, col = "red", lwd=4,)

legend(2,4,legend=c("this (r2 = 0.0249)"," model trained MSE = 27.08677"))
```

```r
# summary 

# in this case, we can find that the corelation of Sales with Tv, radio, and newspaper

# Sales has strong relationship with TV. However, the relationship between sales and radio

# and Newpaper is weak since their mse is so huge. 

# in addition, when test make the prediction, the result will differ to the original one if 

# we input the original dataset, but the difference will be small.
```
